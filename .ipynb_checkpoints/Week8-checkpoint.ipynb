### (b)
$f(x) = (5x^{12} + 2)(\pi - \pi^2 x^4)$

Use product rule: $(fg)' = f'g + fg'$

$f'(x) = 60x^{11}$

$g'(x) = -4\pi^2 x^3$

Apply product rule:

$f'(x)g(x) + f(x)g'(x)$

$(12\pi x^{11} - 10\pi^2 x^{15} - 20\pi^2 x^3)$

### (c)
$f(u) = \sqrt{u} - \frac{3}{u^3} + 2u^7$

Rewrite $\sqrt{u}$ as $u^{1/2}$ and apply power rule:

$f'(u) = \frac{1}{2} u^{-1/2} + 9u^{-4} + 14u^6$

### (d)
$f(x) = mx + b \quad (m, b \text{ are constants})$

Derivative:

$f'(x) = m$

### (e)
$f(x) = 0.5 \sin{x} + \frac{\sqrt[3]{x}}{x} + \pi^2$

Step 1: Derivative of $0.5\sin{x}$ is $0.5\cos{x}$

Step 2: Rewrite $\frac{\sqrt[3]{x}}{x}$ as $x^{-\frac{2}{3}}$

Step 3: Derivative of $x^{-\frac{2}{3}}$ is $-\frac{2}{3}x^{-\frac{5}{3}}$

Step 4: The derivative of a constant $\pi^2$ is 0

Final answer:
$f'(x) = 0.5\cos{x} - \frac{2}{3}x^{-\frac{5}{3}}$

### (f)
$f(x) = \frac{\pi - \pi x^4}{5x^{12} + 2}$

Use quotient rule: $\left(\frac{f}{g}\right)' = \frac{f'g - fg'}{g^2}$

Let $f(x) = \pi - \pi x^4$ and $g(x) = 5x^{12} + 2$

Step 1: $f'(x) = -4\pi x^3$

Step 2: $g'(x) = 60x^{11}$

Apply quotient rule:

$f'(x) = \frac{(-4\pi x^3)(5x^{12} + 2) - (\pi - \pi x^4)(60x^{11})}{(5x^{12} + 2)^2}$

### (g)
$f(x) = 2\sqrt{x} - \frac{1}{\sqrt{x}}$

Rewrite as: $f(x) = 2x^{1/2} - x^{-1/2}$

Step 1: Derivative of $2x^{1/2} is x^{-1/2}$

Step 2: Derivative of $-x^{-1/2} is 0.5x^{-3/2}$

Final answer:

$f'(x) = x^{-1/2} + \frac{1}{2}x^{-3/2}$

### (h)
$f(z) = \tan{z} (\sin{z} - 5)$

Use product rule: $(fg)' = f'g + fg'$

Let $f(z) = \tan{z} and g(z) = \sin{z} - 5$

Step 1: $f'(z) = \sec^2{z}$

Step 2: $g'(z) = \cos{z}$

Apply product rule:

$f'(z) = \sec^2{z}(\sin{z} - 5) + \tan{z}\cos{z}$

### (i)
$f(x) = \frac{\sin{x}}{x^2}$

Use quotient rule: $\left(\frac{f}{g}\right)' = \frac{f'g - fg'}{g^2}$

Step 1: $f'(x) = \cos{x}, g(x) = x^2, g'(x) = 2x$

Apply quotient rule:
$f'(x) = \frac{\cos{x}(x^2) - \sin{x}(2x)}{x^4}$

Final answer:
$f'(x) = \frac{x^2\cos{x} - 2x\sin{x}}{x^4}$

### (j)
$f(x) = x^2 e^x$

Use product rule: $(fg)' = f'g + fg'$

Step 1: $f'(x) = 2x, g'(x) = e^x$

Apply product rule:
$f'(x) = 2xe^x + x^2 e^x$

### (k)
$f(x) = \cos{x} + e^x$

Step 1: Derivative of $\cos{x} is -\sin{x}$

Step 2: Derivative of $e^x is e^x$

Final answer:
$f'(x) = -\sin{x} + e^x$

### (l)
$f(x) = \frac{\sin{x}}{\cos{x}}$

This is the same as $f(x) = \tan{x}$

The derivative of $\tan{x} is \sec^2{x}$

Final answer:
$f'(x) = \sec^2{x}$

### (m)
$f(x) = e^x \ln{\frac{1}{2x}}$

Apply product rule: $(fg)' = f'g + fg'$

Let $f(x) = e^x$ and $g(x) = \ln{\frac{1}{2x}}$

Step 1: $f'(x) = e^x$

Step 2: $g'(x) = \frac{-1}{2x}$

Apply product rule:
$f'(x) = e^x \ln{\frac{1}{2x}} + e^x \frac{-1}{2x}$

### (n)
$f(x) = \frac{10}{10 + \sin{x}}$

Use quotient rule: $\left(\frac{f}{g}\right)' = \frac{f'g - fg'}{g^2}$

Step 1: $f'(x) = 0, g(x) = 10 + \sin{x}, g'(x) = \cos{x}$

Apply quotient rule:

$f'(x) = \frac{0(10 + \sin{x}) - 10\cos{x}}{(10 + \sin{x})^2}$

Final answer:

$f'(x) = \frac{-10\cos{x}}{(10 + \sin{x})^2}$

### (o)
$f(x) = \sin{(e^x \cos{x})}$

Use chain rule: $f'(x) = \cos{u} \cdot u'$

Let $u = e^x \cos{x}$, then $f'(x) = \cos{(e^x \cos{x})} \cdot \frac{d}{dx} (e^x \cos{x})$

Apply product rule to $u = e^x \cos{x}$:

$u'(x) = e^x \cos{x} - e^x \sin{x}$

Final answer:

$f'(x) = \cos{(e^x \cos{x})} \cdot (e^x \cos{x} - e^x \sin{x})$

### (p)
$f(x) = 6e^{\cos{t}} / 5\sqrt[3]{t}$

Step 1: Derivative of $6e^{\cos{t}}$ using chain rule:

$\frac{d}{dt} (6e^{\cos{t}}) = -6e^{\cos{t}} \sin{t}$

Step 2: Derivative of $\frac{1}{5t^{1/3}}$ is $-\frac{1}{15t^{4/3}}$

Final answer:

$f'(t) = \frac{-6e^{\cos{t}} \sin{t}}{5\sqrt[3]{t}} - \frac{2e^{\cos{t}}}{15t^{4/3}}$

### (q)
$f(x) = \ln{(x^2 + xe^x)}$

Use chain rule: $f'(x) = \frac{1}{u} \cdot u'$

Let $u = x^2 + xe^x$

Step 1: Derivative of $u = 2x + e^x + xe^x$

Final answer:

$f'(x) = \frac{2x + e^x + xe^x}{x^2 + xe^x}$

### (r)
$f(x) = \frac{5x^2 + \ln{x}}{7\sqrt{x + 5}}$

Use quotient rule: $\left(\frac{f}{g}\right)' = \frac{f'g - fg'}{g^2}$

Step 1: Derivative of $f(x) = 10x + \frac{1}{x}$

Step 2: Derivative of $g(x) = \frac{7}{2(x+5)^{1/2}}$

Apply quotient rule:

$f'(x) = \frac{(10x + \frac{1}{x})(7\sqrt{x+5}) - (5x^2 + \ln{x})\frac{7}{2(x+5)^{1/2}}}{(7\sqrt{x+5})^2}$

# Activity 2

### (a)
$f(t) + g(t)$

Using the sum rule: $(f + g)' = f'(t) + g'(t)$

$f'(2) = 2, g'(2) = -1$

$f'(2) + g'(2) = 2 + (-1) = 1$

### (b)
$5f(t) - 2g(t)$

Using the constant multiple rule and sum rule:

$f'(t) = 2, g'(t) = -1$

Derivative: $5f'(t) - 2g'(t)$

= $5(2) - 2(-1)$

= $10 + 2 = 12$

### (c)
$f(t)g(t)$

Using the product rule: $(fg)' = f'(t)g(t) + f(t)g'(t)$

$f'(2) = 2, f(2) = 3, g'(2) = -1, g(2) = 4$

$f'(2)g(2) + f(2)g'(2) = 2(4) + 3(-1) = 8 - 3 = 5$

### (d)
$\frac{f(t)}{g(t)}$

Using the quotient rule: $\left(\frac{f}{g}\right)' = \frac{f'(t)g(t) - f(t)g'(t)}{g(t)^2}$

$f'(2) = 2, f(2) = 3, g'(2) = -1, g(2) = 4$

$\frac{f'(2)g(2) - f(2)g'(2)}{g(2)^2} = \frac{2(4) - 3(-1)}{4^2} = \frac{8 + 3}{16} = \frac{11}{16}$

### (e)
$g(f(t))$

Using the chain rule: $(g(f(t)))' = g'(f(t)) \cdot f'(t)$

$f(2) = 3, f'(2) = 2, g'(3) = 17$

$g'(f(2)) \cdot f'(2) = 17 \cdot 2 = 34$

### (f)
$\sqrt{g(t)}$

Using the chain rule: $(\sqrt{g(t)})' = \frac{1}{2\sqrt{g(t)}} \cdot g'(t)$

$g(2) = 4, g'(2) = -1$

$\frac{1}{2\sqrt{4}} \cdot (-1) = \frac{1}{4} \cdot (-1) = -\frac{1}{4}$

### (g)
$t^2 f(t)$

Using the product rule: $(t^2 f(t))' = 2t f(t) + t^2 f'(t)$

$f(2) = 3, f'(2) = 2$

$2(2)(3) + (2^2)(2) = 12 + 8 = 20$

### (h)
$(f(t))^2 + (g(t))^2$

Using the chain rule for both parts: $(f^2)' = 2f(t)f'(t), (g^2)' = 2g(t)g'(t)$

$f(2) = 3, f'(2) = 2, g(2) = 4, g'(2) = -1$

$2f(2)f'(2) + 2g(2)g'(2) = 2(3)(2) + 2(4)(-1) = 12 - 8 = 4$

### (i)
$\frac{1}{f(t)}$

Using the chain rule: $\left(\frac{1}{f(t)}\right)' = -\frac{f'(t)}{f(t)^2}$

$f(2) = 3, f'(2) = 2$

$-\frac{f'(2)}{f(2)^2} = -\frac{2}{3^2} = -\frac{2}{9}$

### (j)
$f(3t - (g(1 + t))^2)$

Using the chain rule and product rule:

Let $u = 3t - (g(1+t))^2$, so $f'(u) = f'(u) \cdot u'$

$u'(t) = 3 - 2g(1+t) \cdot g'(1+t)$

At $t=2: g(1+2) = g(3) = 2, g'(3) = 17$

$u'(2) = 3 - 2(2)(17) = 3 - 68 = -65$

$f'(u) = f'(2), f'(2) = 2$

$f'(u) \cdot u'(2) = 2 \cdot (-65) = -130$

### (k)
To calculate the derivative of $f(g(t))$ at $t = 2$, we would need the value of $g'(2)$ and $f'(g(2))$ to apply the chain rule. We already know $g'(2)$, but we would need $f'(g(2))$.

### (l)
Estimate the value of $\frac{f(t)}{g(t)}$ at $t = 1.95$:

Using linear approximation: $f(1.95) \approx f(2) + f'(2)(1.95 - 2) = 3 + 2(-0.05) = 2.9$

$g(1.95) \approx g(2) + g'(2)(1.95 - 2) = 4 + (-1)(-0.05) = 4.05$

Estimate: $\frac{f(1.95)}{g(1.95)} = \frac{2.9}{4.05} \approx 0.716$


# Activity 3: Pseudo Code Derivative Calculator

```python
1. Start
2. Define function `parse_expression(expression)`:
   - Tokenize the input algebraic expression.
   - Recognize operators (e.g., +, -, *, /) and terms.
   - Return a parsed expression in a structured format like a list or tree.

3. Define function `differentiate_term(term)`:
   - If the term is a constant, return 0.
   - If the term is a variable, return 1.
   - If the term is a power of a variable, use the power rule:
     - Return n * x^(n-1).

4. Define function `differentiate_expression(parsed_expression)`:
   - For each term in the parsed expression:
     - Use `differentiate_term(term)` to compute the derivative of each term.
   - Combine the derivatives of each term to form the final derivative expression.
   - Return the final derivative expression.

5. Define function `simplify_expression(derivative_expression)`:
   - Simplify the result such as combining the like terms.
   - Return the simplified derivative.

6. Define main function `derivative_solver(expression)`:
   - Call `parse_expression(expression)` to parse the input expression.
   - Call `differentiate_expression(parsed_expression)` to differentiate the parsed expression.
   - Call `simplify_expression(derivative_expression)` to simplify the result.
   - Output the final derivative.

7. End
```
