# Activity 1

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function for f(x)
def fnf(x):
    return np.sqrt(1 + x**3)

# Riemann Sum Functions
def left_riemann_sum(a, b, steps, draw_boxes=False):
    delta_x = (b - a) / steps
    x = np.linspace(a, b - delta_x, steps)
    heights = fnf(x)
    areas = heights * delta_x
    total_area = np.sum(areas)
    
    if draw_boxes:
        draw_riemann_boxes(x, heights, delta_x, a, b, 'Left Riemann Sum', 'left')
    
    return total_area

def right_riemann_sum(a, b, steps, draw_boxes=False):
    delta_x = (b - a) / steps
    x = np.linspace(a + delta_x, b, steps)
    heights = fnf(x)
    areas = heights * delta_x
    total_area = np.sum(areas)
    
    if draw_boxes:
        draw_riemann_boxes(x, heights, delta_x, a, b, 'Right Riemann Sum', 'right')
    
    return total_area

def midpoint_riemann_sum(a, b, steps, draw_boxes=False):
    delta_x = (b - a) / steps
    x = np.linspace(a + delta_x/2, b - delta_x/2, steps)
    heights = fnf(x)
    areas = heights * delta_x
    total_area = np.sum(areas)
    
    if draw_boxes:
        draw_riemann_boxes(x, heights, delta_x, a, b, 'Midpoint Riemann Sum', 'center')
    
    return total_area

# Function to draw the Riemann boxes with proper alignment
def draw_riemann_boxes(x, heights, delta_x, a, b, title, alignment):
    plt.figure(figsize=(8, 6))
    
    # Adjust alignment of bars based on the type of Riemann Sum
    if alignment == 'left':
        plt.bar(x, heights, width=delta_x, align='edge', edgecolor='black', alpha=0.5)
    elif alignment == 'right':
        plt.bar(x - delta_x, heights, width=delta_x, align='edge', edgecolor='black', alpha=0.5)
    elif alignment == 'center':
        plt.bar(x - delta_x / 2, heights, width=delta_x, align='edge', edgecolor='black', alpha=0.5)
    
    x_vals = np.linspace(a, b, 1000)
    plt.plot(x_vals, fnf(x_vals), color='red', linewidth=2)
    plt.title(title)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.show()

# Set parameters
a = 1
b = 3
steps = 4  # Adjust this for different number of boxes

# Toggle to draw boxes or not
draw_boxes = True

# Calculate and display results
print("Left Riemann Sum:", left_riemann_sum(a, b, steps, draw_boxes))
print("Right Riemann Sum:", right_riemann_sum(a, b, steps, draw_boxes))
print("Midpoint Riemann Sum:", midpoint_riemann_sum(a, b, steps, draw_boxes))
```

# Activity 2

```python
def fnf(x):
    return np.sqrt(1 + x**3)

# Set parameters for the Riemann sum illustration
a = 1
b = 3
steps = 4  # Adjustable number of boxes

# Calculate delta_x for the intervals
delta_x = (b - a) / steps

# Create x points for left, right, and midpoint Riemann sums
x_left = np.linspace(a, b - delta_x, steps)
x_right = np.linspace(a + delta_x, b, steps)
x_midpoint = np.linspace(a + delta_x / 2, b - delta_x / 2, steps)

# Calculate heights for each method
heights_left = fnf(x_left)
heights_right = fnf(x_right)
heights_midpoint = fnf(x_midpoint)

# Set up the plot
plt.figure(figsize=(12, 8))

# Generate the actual function curve
x_vals = np.linspace(a, b, 1000)
y_vals = fnf(x_vals)
plt.plot(x_vals, y_vals, color='blue', linewidth=2, label='f(x) = sqrt(1 + x^3)')

# Left Riemann Sum (aligned to the left)
plt.bar(x_left, heights_left, width=delta_x, align='edge', edgecolor='black', alpha=0.3, color='red', label='Left Riemann Sum')

# Right Riemann Sum (shifted to align with the right edge)
plt.bar(x_right - delta_x, heights_right, width=delta_x, align='edge', edgecolor='black', alpha=0.3, color='green', label='Right Riemann Sum')

# Midpoint Riemann Sum (centered)
plt.bar(x_midpoint - delta_x / 2, heights_midpoint, width=delta_x, align='edge', edgecolor='black', alpha=0.3, color='orange', label='Midpoint Riemann Sum')

plt.title('Comparison of Left, Right, and Midpoint Riemann Sums')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

The code compares the Left Riemann Sum and Midpoint Riemann Sum values, showing how the midpoint method stabilizes faster, often requiring fewer intervals to achieve a similar level of accuracy as the left endpoint method.

Left Endpoint Riemann Sum:

    Advantages: Simple to calculate, especially when starting from a known initial point. This method works well if the function has a decreasing trend over the interval, as it may provide a closer approximation to the actual integral.
    
    Disadvantages: When the function is increasing, the left endpoint method underestimates the area, leading to less accurate results. Additionally, it can converge more slowly to the actual integral value compared to midpoint or right endpoint sums in cases with non-linear functions
    
Right Endpoint Riemann Sum:

    Advantages: Provides a better estimate for functions that are increasing over the interval. This method can also be helpful for functions with a positive slope, as it captures more of the area.
    
    Disadvantages: For decreasing functions, it tends to overestimate the area, leading to higher errors. Like the left endpoint sum, it may require a higher number of intervals for accurate approximation, especially for functions with curvature.

Midpoint Riemann Sum:

    Advantages: Generally provides the most accurate estimate among the three methods for functions that are either increasing or decreasing, as it balances overestimates and underestimates within each interval. This method often converges faster to the true integral value with fewer intervals, especially in cases of polynomial or oscillatory functions.
    
    Disadvantages: Slightly more complex to calculate as it requires determining the midpoint of each interval. Additionally, if the function has sudden changes in slope, the midpoint method may still introduce errors.

```python
def f(x):
    return np.sqrt(1 + x**3)

# Implement Riemann Sum methods
def left_riemann_sum(f, a, b, n):
    dx = (b - a) / n
    x = np.linspace(a, b - dx, n)
    return np.sum(f(x) * dx)

def right_riemann_sum(f, a, b, n):
    dx = (b - a) / n
    x = np.linspace(a + dx, b, n)
    return np.sum(f(x) * dx)

def midpoint_riemann_sum(f, a, b, n):
    dx = (b - a) / n
    x = np.linspace(a + dx / 2, b - dx / 2, n)
    return np.sum(f(x) * dx)

# Interval and number of intervals
a, b = 1, 3  # The interval over which we are calculating
n_values = [40, 400, 4000, 40000]  # Different numbers of intervals

# Calculate Riemann Sums for each method and each n
results = {
    "n": [],
    "Left Riemann Sum": [],
    "Right Riemann Sum": [],
    "Midpoint Riemann Sum": []
}

for n in n_values:
    results["n"].append(n)
    results["Left Riemann Sum"].append(left_riemann_sum(f, a, b, n))
    results["Right Riemann Sum"].append(right_riemann_sum(f, a, b, n))
    results["Midpoint Riemann Sum"].append(midpoint_riemann_sum(f, a, b, n))

# Create DataFrame
df = pd.DataFrame(results)
print("Riemann Sum Comparison")
print(df)
```

# Activity 4

Trapezoid Rule is ideal when you need a more accurate approximation with minimal increase in complexity.

Simpsons Rule excels for functions with continuous curvature, providing very high accuracy with fewer intervals, making it highly efficient for various scientific applications.

```python
# Function to integrate
def f(x):
    return np.sqrt(1 + x**3)

# Interval
a, b = 1, 3

# Define Riemann, Trapezoid, and Simpson's Rule functions
def left_riemann_sum(f, a, b, n):
    dx = (b - a) / n
    x = np.linspace(a, b - dx, n)
    return np.sum(f(x) * dx)

def right_riemann_sum(f, a, b, n):
    dx = (b - a) / n
    x = np.linspace(a + dx, b, n)
    return np.sum(f(x) * dx)

def trapezoid_rule(f, a, b, n):
    dx = (b - a) / n
    x = np.linspace(a, b, n+1)
    y = f(x)
    return dx * (np.sum(y) - (y[0] + y[-1]) / 2)

def simpsons_rule(f, a, b, n):
    if n % 2 == 1:
        n += 1  # Simpson's rule requires an even number of intervals
    dx = (b - a) / n
    x = np.linspace(a, b, n + 1)
    y = f(x)
    S = dx / 3 * (y[0] + 4 * np.sum(y[1:n:2]) + 2 * np.sum(y[2:n-1:2]) + y[n])
    return S

# Values of n
n_values = [40, 400, 4000, 40000]

# Calculate each sum method and store in DataFrame
results = {
    "n": [],
    "Left Riemann Sum": [],
    "Right Riemann Sum": [],
    "Trapezoid Rule": [],
    "Simpson's Rule": []
}

for n in n_values:
    results["n"].append(n)
    results["Left Riemann Sum"].append(left_riemann_sum(f, a, b, n))
    results["Right Riemann Sum"].append(right_riemann_sum(f, a, b, n))
    results["Trapezoid Rule"].append(trapezoid_rule(f, a, b, n))
    results["Simpson's Rule"].append(simpsons_rule(f, a, b, n))

# Display results in a DataFrame
df_results = pd.DataFrame(results)
print("Integration Method Comparison")
print(df_results)
```

### Trapezoid Rule

By approximating the area under the curve with trapezoids, this method smooths out the steps of traditional Riemann sums, especially for linear or smoothly varying functions.

```python
def f(x):
    return np.sqrt(1 + x**3)

# Define Trapezoid Rule
def trapezoid_rule(f, a, b, n):
    dx = (b - a) / n
    x = np.linspace(a, b, n+1)
    y = f(x)
    return dx * (np.sum(y) - (y[0] + y[-1]) / 2)

# Interval and number of intervals
a, b = 1, 3
n_values = [10, 50, 100, 500, 1000]

# Calculate Trapezoid Rule approximation and actual value for comparison
actual_value = 6.229959  # Approximation from midpoint Riemann sum for reference

results_trap = {
    "n": [],
    "Trapezoid Approximation": [],
    "Error": []
}

for n in n_values:
    approx = trapezoid_rule(f, a, b, n)
    error = abs(actual_value - approx)
    results_trap["n"].append(n)
    results_trap["Trapezoid Approximation"].append(approx)
    results_trap["Error"].append(error)

import pandas as pd
df_trap = pd.DataFrame(results_trap)
print("Trapezoid Rule Approximation and Error")
print(df_trap)
```

### Simpson's Rule
By fitting parabolic segments to the intervals, Simpsons Rule captures more of the functions curvature, resulting in a rapid convergence to the actual value. This makes it highly effective for integrals of functions with continuous second derivatives.

```python
# Define Simpson's Rule (requires an even number of intervals)
def simpsons_rule(f, a, b, n):
    if n % 2 == 1:
        n += 1  # Simpson's rule requires an even number of intervals
    dx = (b - a) / n
    x = np.linspace(a, b, n + 1)
    y = f(x)
    S = dx / 3 * (y[0] + 4 * np.sum(y[1:n:2]) + 2 * np.sum(y[2:n-1:2]) + y[n])
    return S

# Calculate Simpson's Rule approximation and error for comparison
results_simpson = {
    "n": [],
    "Simpson's Approximation": [],
    "Error": []
}

for n in n_values:
    approx = simpsons_rule(f, a, b, n)
    error = abs(actual_value - approx)
    results_simpson["n"].append(n)
    results_simpson["Simpson's Approximation"].append(approx)
    results_simpson["Error"].append(error)

df_simpson = pd.DataFrame(results_simpson)
print("Simpson's Rule Approximation and Error")
print(df_simpson)
```
