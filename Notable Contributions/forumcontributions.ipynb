### Week 8

Derivative Solving Program Pseudocode:

1. Start
2. Define function `parse_expression(expression)`:
   - Tokenize the input algebraic expression.
   - Recognize operators (e.g., +, -, *, /) and terms.
   - Return a parsed expression in a structured format like a list or tree.

3. Define function `differentiate_term(term)`:
   - If the term is a constant, return 0.
   - If the term is a variable, return 1.
   - If the term is a power of a variable, use the power rule:
     - Return n * x^(n-1).

4. Define function `differentiate_expression(parsed_expression)`:
   - For each term in the parsed expression:
     - Use `differentiate_term(term)` to compute the derivative of each term.
   - Combine the derivatives of each term to form the final derivative expression.
   - Return the final derivative expression.

5. Define function `simplify_expression(derivative_expression)`:
   - Simplify the result such as combining the like terms.
   - Return the simplified derivative.

6. Define main function `derivative_solver(expression)`:
   - Call `parse_expression(expression)` to parse the input expression.
   - Call `differentiate_expression(parsed_expression)` to differentiate the parsed expression.
   - Call `simplify_expression(derivative_expression)` to simplify the result.
   - Output the final derivative.

7. End
