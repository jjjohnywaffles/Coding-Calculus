```python
import math

# Define the function fnf(t)
def fnf(t):
    return math.cos(t ** 2)

t_initial = 0   # Initial time
t_final = 4     # Final time
number_of_steps = 2 ** 3  # Number of steps: 8
delta_t = (t_final - t_initial) / number_of_steps  # Step size: how much t increases per iteration

# starting values
t = t_initial  # Start time
accumulation = 0  # Start with no accumulation

# Loop over the number of steps to calculate deltay, accumulation, and update t
for k in range(1, number_of_steps + 1):
    deltay = fnf(t) * delta_t  # Calculate the delta y 
    accumulation += deltay  # Add deltay to the running total (accumulation)
    t += delta_t  
    
    print(f"Step {k}: deltay = {deltay:.4f}, accumulation = {accumulation:.4f}, t = {t:.4f}")
```

### Purpose: 

This code calculates the accumulation of a function (cos(t^2)) over a series of small intervals between an initial time (t_initial = 0) and a final time (t_final = 4).
    
### Key Steps:

The interval between t_initial and t_final is broken into a specified number of steps (number_of_steps = 8 in this case), and for each step, the value of the function is calculated and "accumulated" over time.

### deltay:

This represents the small change in the function's value at each step, calculated as fnf(t) * delta_t.

### Accumulation:

This keeps a running total of all the small deltay values, which represents the accumulated effect of the function over the time interval.

### Output:

For each step, it prints the current deltay, the accumulated total so far, and the current time t.

### Demo:

    The code breaks the interval from 0 to 4 into 8 steps.
    At each step, it calculates a small portion of the function cos(t^2), multiplies it by the step size (delta_t), and adds it to the accumulated total.
    This is essentially an approximation of the integral of the function over the interval using numerical methods.

```python
import matplotlib.pyplot as plt
import math

def fnf(t):
    return math.cos(t ** 2)

# Set initial and final times
t_initial = 0
t_final = 10

# Set the number of steps
number_of_steps = 400 

# Calculate the step size (delta t)
delta_t = (t_final - t_initial) / number_of_steps

# Initialize starting time and accumulation
t = t_initial
accumulation = 0

# Lists to store data for plotting
t_values = []
accumulation_values = []

# Loop through each step to calculate accumulation and plot the values
for k in range(1, number_of_steps + 1):
    # Calculate the small change in function value (deltay)
    deltay = fnf(t) * delta_t
    
    # Update the accumulation by adding the small change (deltay)
    accumulation += deltay
    
    # Store the current time (t) and accumulation for plotting
    t_values.append(t)
    accumulation_values.append(accumulation)
    
    t += delta_t

plt.plot(t_values, accumulation_values, label="Accumulation Function")
plt.xlabel('Time (t)')
plt.ylabel('Accumulation')
plt.title('Accumulation Function over Time')
plt.grid(True)
plt.legend()
plt.show()
```

### Purpose:

    This program numerically integrates the function cos⁡(t2) over a specified interval (from t=0 to t=4) and plots the accumulation (total sum of the function) over time.

### Explanation:

    The program divides the time interval into a large number of small steps (number_of_steps = 400), which makes the plot smooth.
    For each step, it calculates the small change in the function (deltay) and adds it to the total accumulation.
    The result is a plot showing how the total accumulation of cos⁡(t2) increases over time.

### Key Parts:

    fnf(t): The function cos⁡(t2) is defined.
    delta_t: The time interval for each small step is calculated.
    accumulation: This variable keeps track of the total sum of the function over time.
    t_values and accumulation_values: Lists to store the time and accumulation values for plotting.

```python
import matplotlib.pyplot as plt
import math

# Define each of the six functions
def fnf_1(t):
    return (827 / (10.5 ** 2 + (t - 55) ** 2))

def fnf_2(t):
    return (1400 / (10 ** 2 + (t - 45) ** 2))

def fnf_3(t):
    return (1200 / (10 ** 2 + (t - 48) ** 2))

def fnf_4(t):
    return (827 / (10.5 ** 2 + (t - 44) ** 4))

def fnf_5(t):
    return (1000 / (10.5 ** 2 + (t - 49) ** 2))

def fnf_6(t):
    return (827 / (10.5 ** 2 + (t - 48) ** 2))

def calculate_accumulation(fnf):
    t_initial = 0
    t_final = 89
    number_of_steps = 400
    delta_t = (t_final - t_initial) / number_of_steps
    t = t_initial
    accumulation = 0
    t_values = []
    accumulation_values = []
    
    for k in range(1, number_of_steps + 1):
        deltay = fnf(t) * delta_t
        accumulation += deltay
        t_values.append(t)
        accumulation_values.append(accumulation)
        t += delta_t
    
    return t_values, accumulation_values

functions = [fnf_1, fnf_2, fnf_3, fnf_4, fnf_5, fnf_6]
labels = ["A", "B", "C", "D", "E", "F"]

for fnf, label in zip(functions, labels):
    t_values, accumulation_values = calculate_accumulation(fnf)
    plt.plot(t_values, accumulation_values, label=label)

plt.xlabel('Time (t)')
plt.ylabel('Accumulation')
plt.title('Accumulation Functions')
plt.grid(True)
plt.legend()
plt.show()
```
