```python
def lynxHare(num_hare, num_lynx, t_final):
    # time variables
    t_initial = 0
    t = t_initial
    # Starting values at the beginning of model for hares and lynx (turns out the plural of lynx is just lynx)
    H = num_hare
    L = num_lynx

    # Hares per month per hare
    a = .1
    # Hares per month per hare-lynx
    b = .005
    # Lynx per month per hare-lynx
    c = .00004
    # Lynx per month per lynx
    d = .04

    # Creating empty arrays to hold values we will plot
    H_graph = []
    L_graph = []

    # Change in time/number of steps
    delta_t = 1
    # Iterating through the number of line segments in our plot (the more lines, the smoother the graph)
    for k in range (1, t_final):
        # Add current values to graph vectors (include Lynx scaling factor of 60)
        H_graph.append(H)
        L_graph.append(60*L)
        # Rate of change in Hare population
        Hprime = a*H - b*H*L
        # Rate of change in Lynx population
        Lprime = c*H*L - d*L
        
        # Actual change in Hare and Lynx populations
        delta_H = Hprime*delta_t
        delta_L = Lprime*delta_t

        #Updating the values of t, H, and L with our delta values
        t = t + delta_t
        H = H + delta_H
        L = L + delta_L

    # return the arrays with values over each step
    return H_graph, L_graph


# Plot the populations
H_line, L_line= lynxHare(2000, 10, 250)
plt.plot(H_line, color='blue', label='Hares')
plt.plot(L_line, color='orange', label='Lynx')
plt.legend(['Hares', 'Lynx'])
plt.xlabel('Months')
plt.ylabel('Population (1 Hare, 60 Lynx)')
plt.title('Hare and Lynx Populations')    
```

The graph above shows part C of number 7 on p. 193-194 of Calculus in Context, beginning with 2,000 hares and 10 lynx.  It includes the basic assumptions of the two species model:

* In the absence of lynx, the hare population grows logistically - we see this in the equation for Hprime, where if the lynx population is 0 then the hare population is aH 
* The population of hares declines in proportion to to H*L - see this in the second half of the equation for Hprime, -bHL.  Both of these assumptions can be seen in the rise and fall of the blue line, which represents the hare population
* In the absence of hares, the lynx die of proportional to the number of lynx - the equation for Lprime shows that if H is 0, then the lynx decline at a rate of dL (and is demonstrated by the declines in the orange line, which occur following the decline of the blue line representing hares)
* The lynx population increases proportionately to the number of encounters between lynx and hares - this can be seen in the first half of Lprime, cHL where c represents lynx per month per hare-lynx.

The hare-lynx units are especially interesting, as they complicate the model to include not only the populations of predators and prey, but also the encounters between the two (R.I.P. hares).

# a)
D(x) suggests a saturation process where the rate at which a single fox eats rabbits increases with the number of rabbits x but approaches a limit as x increases. This type of function is the maximum rate at which a fox can eat rabbits. The parameter c represents the maximum number of rabbits a fox could eat if there were an unlimited supply of rabbits. The parameter d acts as a half-saturation constant: the number of rabbits at which the fox's consumption rate is half its maximum. Making d smaller or larger affects how quickly the consumption rate approaches its maximum. A smaller d means the rate reaches near-maximum at a lower rabbit population.

```python
# sketch of part a)

def D(x, c, d):
    return (c * x) / (x + d)

# Generate x values from 0 to 100
x_values = np.linspace(0, 100, 400)
c = 10  # Max consumption rate
d = 20  # Half-saturation constant

# Compute D(x) for each x
D_values = D(x_values, c, d)

# Plotting the function
plt.figure(figsize=(8, 5))
plt.plot(x_values, D_values, label=f'D(x) = {c}x / (x + {d})', color='blue')
plt.title('Graph of D(x) = cx / (x + d)')
plt.xlabel('Number of Rabbits (x)')
plt.ylabel('Rate of Rabbit Consumption by a Fox (D(x))')
plt.axhline(y=c, color='r', linestyle='--', label='Saturation value c')
plt.legend()
plt.grid(True)
plt.show()
```

# b)
These system of equations incorporate May's Assumption by including the following:
### 1) 
Logistic growth of rabbits in the absence of foxes: a * x(1-(x/b))
### 2)
Decrease in rabbit population due to predation by foxes, which is moderated by the function D(x) = cxy/(x+b)
### 3)
Logistic growth of the fox population, which depends on the rabbit population providing enough food to sustain it: e * y(1-y/(fx))

# c) and d)

### c) 
The rabbit and fox populations show oscillatory behavior initially, with periods of growth and decline. The rabbit population grows rapidly due to logistic growth when foxes are few. As the fox population catches up, it preys on the rabbits, reducing their numbers. Over time, the amplitude of oscillations slightly decreases, and the populations of both species stabilize around equilibrium values. This behavior suggests a stable coexistence between the fox and rabbit populations in the long term.

### d)
The different initial conditions reveal that higher initial fox populations lead to a rapid decline in rabbits, which in turn causes a drop in fox numbers due to lack of food, while lower initial rabbit populations result in slower growth for both species. In all cases, both populations stabilize over time, with lower initial rabbit or higher initial fox populations leading to a lower overall equilibrium for both species.

May's Model vs. Lotka-Volterra: The key difference lies in the saturation behavior and bounded growth incorporated in May's model. While the Lotka-Volterra model shows continuous oscillations, May's model tends to stabilize over time due to the more realistic assumptions of predation saturation and the carrying capacity of the prey.

```python
# code for part c and d

from scipy.integrate import solve_ivp

a = 0.6  # Growth rate of rabbits
b = 10   # Carrying capacity of rabbits
c = 0.5  # Rate of predation
d = 1    # Half-saturation constant
e = 0.1  # Growth rate of foxes
f = 2    # Effect of rabbits on carrying capacity of foxes

# Equations for May's model
def may_model(t, z):
    x, y = z
    dxdt = a * x * (1 - x / b) - (c * x * y) / (x + d)
    dydt = e * y * (1 - y / (f * x))
    return [dxdt, dydt]

def solve_and_plot(initial_conditions, t_span, t_eval):
    solution = solve_ivp(may_model, t_span, initial_conditions, t_eval=t_eval, method='RK45')

    # Plot the results
    plt.figure(figsize=(10, 6))
    plt.plot(solution.t, solution.y[0], label='Rabbits (x)', color='blue')
    plt.plot(solution.t, solution.y[1], label='Foxes (y)', color='orange')
    plt.title(f'May Model with Initial Conditions: Rabbits={initial_conditions[0]*100}, Foxes={initial_conditions[1]}')
    plt.xlabel('Time (months)')
    plt.ylabel('Population')
    plt.legend()
    plt.grid(True)
    plt.show()

# Time span for the simulation
t_span = (0, 200)  # Simulate for 200 months
t_eval = np.linspace(t_span[0], t_span[1], 1000)

# Part c: Initial condition: 20 hectorabbits (2000 rabbits) and 10 foxes
solve_and_plot([20, 10], t_span, t_eval)

# Part d: Different initial conditions
# Case 1: 20 hectorabbits (2000 rabbits) and 20 foxes
solve_and_plot([20, 20], t_span, t_eval)

# Case 2: 10 hectorabbits (1000 rabbits) and 10 foxes
solve_and_plot([10, 10], t_span, t_eval)

# Case 3: 10 hectorabbits (1000 rabbits) and 20 foxes
solve_and_plot([10, 20], t_span, t_eval)
```

# e)
As the parameter c is varied from 0.5 to 0.1, the model exhibits different behaviors in the dynamics of the rabbit and fox populations. For higher values of c, closer to 0.5, the foxes can consume rabbits at a higher rate, which leads to pronounced oscillations in both populations. As c decreases, the saturation effect of predation becomes more pronounced, meaning foxes are less efficient at consuming rabbits even when they are abundant. Consequently, the amplitude of oscillations in the populations reduces, and the system tends to stabilize. The qualitative change occurs between c=0.3 and c=0.25, where a Hopf bifurcation is observed. The system shifts from exhibiting sustained oscillatory cycles to reaching a stable equilibrium where both populations coexist without further oscillations.

```python
def solve_and_plot_c_variation(c_values, t_span, t_eval, initial_conditions):
    plt.figure(figsize=(14, 8))

    # Loop through different values of c
    for c in c_values:
        def may_model_varying_c(t, z):
            x, y = z
            dxdt = a * x * (1 - x / b) - (c * x * y) / (x + d)
            dydt = e * y * (1 - y / (f * x))
            return [dxdt, dydt]

        solution = solve_ivp(may_model_varying_c, t_span, initial_conditions, t_eval=t_eval, method='RK45')
        plt.plot(solution.t, solution.y[0], label=f'Rabbits (c={c})', alpha=0.7)
        plt.plot(solution.t, solution.y[1], '--', label=f'Foxes (c={c})', alpha=0.7)

    plt.title('Effect of Varying c on Rabbit and Fox Populations')
    plt.xlabel('Time (months)')
    plt.ylabel('Population')
    plt.legend()
    plt.grid(True)
    plt.show()

# Range of c values to test
c_values = np.arange(0.5, 0.0, -0.05)
solve_and_plot_c_variation(c_values, t_span, t_eval, [20, 10])  # Initial conditions: 2000 rabbits and 10 foxes
```

### a)
$\frac{dA}{dt} = 0.05Y(t)$

### b)

    Y(t) is the yeast population at time t.
    A(t) is the alcohol amount at time t.
    The term 0.2 Y (1 - Y/10) represents the logistic growth of yeast.
    The term -0.1 Y A represents the reduction in yeast population due to alcohol toxicity.
    
$\frac{dY}{dt} = 0.2 Y \left(1 - \frac{Y}{10}\right) - 0.1 Y A$

### c)

```python
# Parameters
r = 0.2    # Growth rate of yeast
K = 10     # Carrying capacity in pounds
alpha = 0.1  # Toxicity effect (alcohol killing yeast)
beta = 0.05  # Alcohol production rate per yeast

# Differential equations
def yeast_alcohol_system(t, Z):
    Y, A = Z
    dYdt = r * Y * (1 - Y / K) - alpha * Y * A
    dAdt = beta * Y
    return [dYdt, dAdt]

# Initial conditions
Y0 = 0.5  # Initial yeast population in pounds
A0 = 0    # Initial alcohol amount in pounds
initial_conditions = [Y0, A0]

# Time span for simulation
t_span = (0, 200)  
t_eval = np.linspace(t_span[0], t_span[1], 1000)  # For smooth curve

solution = solve_ivp(yeast_alcohol_system, t_span, initial_conditions, t_eval=t_eval, method='RK45')

Y_values = solution.y[0]
A_values = solution.y[1]
t_values = solution.t

plt.figure(figsize=(10, 6))
plt.plot(t_values, Y_values, label='Yeast Population (Y)')
plt.plot(t_values, A_values, label='Alcohol Amount (A)')
plt.axhline(y=0.01, color='red', linestyle='--', label='Yeast threshold (0.01 lbs)')
plt.title('Yeast Population and Alcohol Production Over Time')
plt.xlabel('Time (hours)')
plt.ylabel('Amount (lbs)')
plt.grid(True)
plt.legend()
plt.show()

# Identify when yeast population drops below 0.01 lbs
end_fermentation_time = t_values[np.argmax(Y_values <= 0.01)]
end_alcohol_amount = A_values[np.argmax(Y_values <= 0.01)]

print(f"Fermentation ends at approximately {end_fermentation_time:.2f} hours.")
print(f"At this time, the amount of alcohol produced is approximately {end_alcohol_amount:.2f} lbs.")
```

# #11)

### a)
Let S(t) be the amount of sugar at time (t).
The rate at which sugar is consumed depends on the current yeast population Y(t).
Given that yeast consumes 0.15 lbs of sugar per hour per pound of yeast:

$\frac{dS}{dt} = -0.15 Y$

### b)
The carrying capacity (K) of yeast now depends on the amount of sugar (S).

$K(S) = 0.4 S$

To maintain a carrying capacity of 10 lbs of yeast:

$K(S) = 10 = 0.4 S \implies S = \frac{10}{0.4} = 25 \text{ lbs of sugar}$

To maintain a carrying capacity of 1 lb of yeast:

$K(S) = 1 = 0.4 S \implies S = \frac{1}{0.4} = 2.5 \text{ lbs of sugar}$


The new logistic equation for the growth of yeast Y(t), considering the variable carrying capacity and the toxic effect of alcohol A(t), is:

$\frac{dY}{dt} = r Y \left(1 - \frac{Y}{0.4 S}\right) - \alpha Y A$

where:
r = 0.2: natural growth rate of yeast.
$\alpha$ = 0.1 : rate at which alcohol kills yeast.

### c)
1. **Yeast growth** (logistic with sugar-dependent carrying capacity and alcohol toxicity):

$\frac{dY}{dt} = 0.2 Y \left(1 - \frac{Y}{0.4 S}\right) - 0.1 Y A$

2. **Alcohol production** (proportional to the yeast population):

$\frac{dA}{dt} = 0.05 Y$

3. **Sugar consumption** (proportional to the yeast population):

$\frac{dS}{dt} = -0.15 Y$


```python
r = 0.2       # Growth rate of yeast
alpha = 0.1   # Toxicity effect (alcohol killing yeast)
beta = 0.05   # Alcohol production rate per yeast
gamma = 0.15  # Sugar consumption rate per yeast

def yeast_alcohol_sugar_system(t, Z):
    Y, A, S = Z
    K = 0.4 * S  # Carrying capacity as a function of sugar
    dYdt = r * Y * (1 - Y / K) - alpha * Y * A
    dAdt = beta * Y
    dSdt = -gamma * Y
    return [dYdt, dAdt, dSdt]

# Initial conditions
Y0 = 0.5  # Initial yeast population in pounds
A0 = 0    # Initial alcohol amount in pounds
S0 = 25   # Initial sugar amount in pounds
initial_conditions = [Y0, A0, S0]

t_span = (0, 200)  
t_eval = np.linspace(t_span[0], t_span[1], 1000) 

solution = solve_ivp(yeast_alcohol_sugar_system, t_span, initial_conditions, t_eval=t_eval, method='RK45')

Y_values = solution.y[0]
A_values = solution.y[1]
S_values = solution.y[2]
t_values = solution.t

plt.figure(figsize=(12, 6))
plt.plot(t_values, Y_values, label='Yeast Population (Y)')
plt.plot(t_values, A_values, label='Alcohol Amount (A)')
plt.plot(t_values, S_values, label='Sugar Amount (S)')
plt.axhline(y=0.01, color='red', linestyle='--', label='Yeast threshold (0.01 lbs)')
plt.title('Yeast, Alcohol, and Sugar Over Time')
plt.xlabel('Time (hours)')
plt.ylabel('Amount (lbs)')
plt.grid(True)
plt.legend()
plt.show()

# Identify when yeast population drops below 0.01 lbs
end_fermentation_time = t_values[np.argmax(Y_values <= 0.01)]
end_alcohol_amount = A_values[np.argmax(Y_values <= 0.01)]
end_sugar_amount = S_values[np.argmax(Y_values <= 0.01)]

print(f"Fermentation ends at approximately {end_fermentation_time:.2f} hours.")
print(f"At this time, the amount of alcohol produced is approximately {end_alcohol_amount:.2f} lbs.")
print(f"The amount of sugar left is approximately {end_sugar_amount:.2f} lbs.")
```

# Expanding the Lotka-Volterra Model to a Complex Ecosystem: Mice, Foxes, Bears, Aliens, and a Zombie Apocalypse

The goal of this Lotka-Volterra model modification is to develop a multi-species ecological model that builds upon the Lotka-Volterra predator-prey equations. Each species will have unique interactions based on their roles as predators, prey, or special creatures like zombies and aliens. We can observe how these interactions shape the dynamics of the ecosystem.

Species and Interactions

    Mice (Prey)
        Growth: Mice reproduce quickly.
        Predators: Eaten by foxes and bears.
        Zombies: Can turn mice into zombies upon interaction.

    Foxes (Predator of Mice)
        Predators: Eaten by bears and aliens.
        Prey: Feed on mice.
        Competition: Compete with bears for food.

    Bears (Predator of Mice and Foxes)
        Prey: Feed on mice and foxes.
        Predators: Avoided by aliens but can be turned into zombies.
        Slow Reproduction: Longer lifespan but slower reproduction.

    Aliens (Super-Predators)
        Predators: Hunt foxes and bears.
        Special Ability: Introduce chaos into the ecosystem by occasionally abducting random creatures.

    Zombies (Infectious Species)
        Predators: Infect all other species, converting them into zombies.
        No Reproduction: Zombies do not reproduce in a biological sense but spread by infection.

Model Equations:

    Mice: dMdt = alpha * M - beta_FM * F * M - beta_BM * B * M - beta_ZM * Z * M
    Foxes: dFdt = delta_FM * F * M - gamma_F * F - beta_BF * B * F - beta_AF * A * F - beta_ZF * Z * F
    Bears: dBdt = delta_BM * B * M + delta_BF * B * F - gamma_B * B - beta_ZB * Z * B
    Aliens: dAdt = delta_AF * A * F - gamma_A * A
    Zombies: dZdt = beta_ZM * Z * M + beta_ZF * Z * F + beta_ZB * Z * B - gamma_Z * Z

```python
# Define parameters, these are randomly selected values, these do not reflect a real world rates.
alpha = 0.5  # Growth rate of mice
beta_FM = 0.01  # Fox predation rate on mice
beta_BM = 0.005  # Bear predation rate on mice
beta_ZM = 0.01  # Zombie infection rate on mice
delta_FM = 0.01  # Fox reproduction rate from mice
gamma_F = 0.1  # Death rate of foxes
beta_BF = 0.005  # Bear predation rate on foxes
beta_AF = 0.1  # Alien predation rate on foxes
beta_ZF = 0.02  # Zombie infection rate on foxes
delta_BM = 0.005  # Bear reproduction rate from mice
delta_BF = 0.002  # Bear reproduction rate from foxes
gamma_B = 0.05  # Death rate of bears
beta_ZB = 0.01  # Zombie infection rate on bears
delta_AF = 0.02  # Alien reproduction rate from foxes
gamma_A = 0.05  # Death rate of aliens
gamma_Z = 0.5  # Death rate of zombies

# Initial populations
M0, F0, B0, A0, Z0 = 500, 30, 10, 5, 1

# System of differential equations
def ecosystem(t, y):
    M, F, B, A, Z = y
    dMdt = alpha * M - beta_FM * F * M - beta_BM * B * M - beta_ZM * Z * M
    dFdt = delta_FM * F * M - gamma_F * F - beta_BF * B * F - beta_AF * A * F - beta_ZF * Z * F
    dBdt = delta_BM * B * M + delta_BF * B * F - gamma_B * B - beta_ZB * Z * B
    dAdt = delta_AF * A * F - gamma_A * A
    dZdt = beta_ZM * Z * M + beta_ZF * Z * F + beta_ZB * Z * B - gamma_Z * Z
    return [dMdt, dFdt, dBdt, dAdt, dZdt]

# Time span for simulation
t_span = (0, 5)
t_eval = np.linspace(t_span[0], t_span[1], 1000)

# Initial conditions
initial_conditions = [M0, F0, B0, A0, Z0]

# Solve the system using solve_ivp
solution = solve_ivp(ecosystem, t_span, initial_conditions, t_eval=t_eval)

# Extract solutions
M_values, F_values, B_values, A_values, Z_values = solution.y

# Plot results
plt.figure(figsize=(10, 6))
plt.plot(t_eval, M_values, label='Mice')
plt.plot(t_eval, F_values, label='Foxes')
plt.plot(t_eval, B_values, label='Bears')
plt.plot(t_eval, A_values, label='Aliens')
plt.plot(t_eval, Z_values, label='Zombies')
plt.xlabel('Time')
plt.ylabel('Population')
plt.title('Expanded Lotka-Volterra Ecosystem Dynamics')
plt.legend()
plt.grid(True)
plt.show()
```
