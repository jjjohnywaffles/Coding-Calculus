```python
def Numerical_Point_Derivative(f, a, method='central', h=1e-5):
    if method == 'central':
        # Central difference method (Q1)
        return (f(a + h) - f(a - h)) / (2 * h)
    elif method == 'forward':
        # Forward difference method (Q2)
        return (f(a + h) - f(a)) / h
    else:
        raise ValueError("Method must be 'central' or 'forward'")

# Function Input
def f(x):
    return 2**x

# Test the function
a = 5  # The point at which to find the derivative (x)
# The Central Difference Method is generally preferred for its higher accuracy compared to the forward difference method, 
# especially when a sufficiently small step size h is used
print("Central Difference Method:", Numerical_Point_Derivative(f, a, method='central'))
print("Forward Difference Method:", Numerical_Point_Derivative(f, a, method='forward'))
```

```python
import numpy as np
import matplotlib.pyplot as plt

def Numerical_Function_Derivative(f, x_start, x_end, num_points, method='central'):
    x_values = np.linspace(x_start, x_end, num_points)
    derivative_values = [Numerical_Point_Derivative(f, x, method) for x in x_values]

    plt.figure(figsize=(10, 5))
    plt.plot(x_values, derivative_values, label=f"Derivative using {method} method")
    plt.xlabel('x')
    plt.ylabel("f'(x)")
    plt.title('Numerical Derivative of the Function')
    plt.legend()
    plt.grid(True)
    plt.show()

def f(x):
    return 2**x  # Function: f(x) = 2^x

# Call the function to plot the derivative from x = 1 to x = 3
Numerical_Function_Derivative(f, 1, 3, 20)
```

# #5

From the graphs, 20 points seems to be a reasonable amount to achieve adequate "smoothness" in the graph. However, the numerical results should be fairly accurate for smoother functions but may deviate for functions with faster changing derivatives. For these derivatives, 20 points may not be enough and will need more points to better approximate the derivative in those cases. 

```python
def f1(x):
    return 1/x

def df1(x):
    return -1/x**2

def f2(x):
    return np.sin(7*x)

def df2(x):
    return 7*np.cos(7*x)

def f3(x):
    return x**3

def df3(x):
    return 3*x**2

def f4(x):
    return 2**x

def df4(x):
    return np.log(2) * 2**x

def plot_function_and_derivative(f, df, x_start, x_end, num_points):
    x_values = np.linspace(x_start, x_end, num_points)
    actual_derivative_values = [df(x) for x in x_values]
    numerical_derivative_values = [Numerical_Point_Derivative(f, x) for x in x_values]

    plt.figure(figsize=(10, 5))
    plt.plot(x_values, actual_derivative_values, label='Actual f\'(x)', linestyle='--')
    plt.plot(x_values, numerical_derivative_values, label='Numerical f\'(x)', marker='o', linestyle='')
    plt.xlabel('x')
    plt.ylabel("f'(x)")
    plt.title(f'Comparison of Actual and Numerical Derivatives')
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot derivatives for each function
plot_function_and_derivative(f1, df1, 1, 3, 20)
plot_function_and_derivative(f2, df2, 1, 3, 20)
plot_function_and_derivative(f3, df3, 1, 3, 20)
plot_function_and_derivative(f4, df4, 1, 3, 20)
```
